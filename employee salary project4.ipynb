```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,mean_squared_error
```

```python
data=pd.read_csv('empl.csv')
df=pd.DataFrame(data=data)
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.heatmap(df.isnull())
```

```python
from sklearn.impute import SimpleImputer
```

```python
si=SimpleImputer(strategy="most_frequent")
```

```python
df['City']=si.fit_transform(df['City'].values.reshape(-1,1))
```

```python
df.head()
```

```python
df['Age']=si.fit_transform(df['Age'].values.reshape(-1,1))
```

```python
df.head()
```

```python
df
```

```python
df['Salary']=si.fit_transform(df['Salary'].values.reshape(-1,1))
```

```python
df
```

```python
sns.heatmap(df.isnull())
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
df['City']=le.fit_transform(df['City'])
```

```python
df
```

```python
df['Country']=le.fit_transform(df['Country'])
```

```python
df
```

```python
dfcopy=df.copy()
```

```python
dfcopy.drop("SNo",axis=1,inplace=True)
```

```python
dfcopy.drop("Name",axis=1,inplace=True)
```

```python
dfcopy
```

```python
x=dfcopy.iloc[:,:-1].values
y=dfcopy.iloc[:,-1].values
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.33,random_state=47)
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x_train,y_train)
```

```python
pred=lr.predict(x_test)
```

```python
mean_squared_error(y_test,pred)
```

```python
r2_score(y_test,pred)
```

```python
td=np.array([30.0,9,4])
td=td.reshape(1,-1)
lr.predict(td)
```
